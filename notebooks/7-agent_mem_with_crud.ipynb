```python
from ai.agents import get_document_agent
from langgraph.checkpoint.memory import InMemorySaver
import uuid

checkpointer = InMemorySaver()


config = {
    "configurable": {
        "user_id": 2,
        "thread_id": uuid.uuid4(),
    }
}


# google_doc_agent = get_document_agent(
#     model_required={"provider": "Google", "model": "gemini-2.0-flash"},
#     checkpointer=checkpointer,
# )

openAI_doc_agent = get_document_agent(
    model_required={"provider": "OpenAI", "model": "gpt-5-nano"},
    checkpointer=checkpointer,
)
```

```python


ai_response = openAI_doc_agent.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": "Give me a summary of the movie: Mission impossible 1",
            }
        ]
    },
    config=config,
)

ai_response["messages"][-1].content
```

```python
ai_response = openAI_doc_agent.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": "Great, Now save it to my documents",
            }
        ]
    },
    config=config,
)

ai_response["messages"][-1].content
```
