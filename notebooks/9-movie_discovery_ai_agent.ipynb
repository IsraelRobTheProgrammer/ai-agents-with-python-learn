```python
from ai.agents import get_movie_discovery_agent
from langgraph.checkpoint.memory import InMemorySaver
import uuid

checkpointer = InMemorySaver()


config = {
    "configurable": {
        "user_id": 2,
        "thread_id": uuid.uuid4(),
    }
}


# google_doc_agent = get_movie_agent(
#     model_required={"provider": "Google", "model": "gemini-2.0-flash"},
#     checkpointer=checkpointer,
# )

openAI_movie_agent = get_movie_discovery_agent(
    model_required={"provider": "OpenAI", "model": "gpt-5-nano"},
    checkpointer=checkpointer,
)
```

```python


ai_response = openAI_movie_agent.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": "Tell me about the movie Avengers Endgame",
            }
        ]
    },
    config=config,
)
```
